```python
# üìù Homework / Lab
# Classify complexity of these tasks:

# Checking if a list has duplicates. (O(n) or O(n log n) depending on approach)
# Sorting a dataset of 1M entries. (O(n log n) for efficient algorithms)
# Training k-NN on n samples. (O(n*d) or O(n^2) for training depending on implementation and distance metric)
# Running BFS on a graph with V vertices, E edges. (O(V + E))
# Write your own binary search and prove its Œò(log n). (O(log n))

# Compare runtime of:

# fib_recur(n) vs an iterative Fibonacci implementation.
# Plot time taken for n = 10‚Ä¶40

import time
import matplotlib.pyplot as plt

# Recursive Fibonacci
def fib_recur(n):
    if n <= 1:
        return n
    return fib_recur(n-1) + fib_recur(n-2)

# Iterative Fibonacci
def fib_iter(n):
    a, b = 0, 1
    for _ in range(n):
        a, b = b, a+b
    return a

n_values = range(10, 41)
recursive_times = []
iterative_times = []

print("Comparing Fibonacci runtimes...")

for n in n_values:
    start_time = time.time()
    fib_recur(n)
    end_time = time.time()
    recursive_times.append(end_time - start_time)
    print(f"Recursive Fibonacci({n}): {end_time - start_time:.6f} seconds")

    start_time = time.time()
    fib_iter(n)
    end_time = time.time()
    iterative_times.append(end_time - start_time)
    print(f"Iterative Fibonacci({n}): {end_time - start_time:.6f} seconds")

plt.figure(figsize=(10, 6))
plt.plot(n_values, recursive_times, label='Recursive Fibonacci')
plt.plot(n_values, iterative_times, label='Iterative Fibonacci')
plt.xlabel('n')
plt.ylabel('Time (seconds)')
plt.title('Runtime Comparison of Recursive and Iterative Fibonacci')
plt.legend()
plt.grid(True)
plt.show()
```
